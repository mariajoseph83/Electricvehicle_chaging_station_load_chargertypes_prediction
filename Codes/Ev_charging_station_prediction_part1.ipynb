```python
import pandas as pd
import random

# List of Indian states
indian_states = [
    "A & N Islands", "Andhra Pradesh", "Arunachal Pradesh", "Assam", "Bihar",
    "Chandigarh", "Chhattisgarh", "Dadra & Nagar Haveli and Daman & Diu", "Delhi",
    "Goa*", "Gujarat", "Haryana", "Himachal Pradesh", "Jammu Kashmir", "Jharkhand",
    "Karnataka", "Kerala", "Ladakh", "Lakshadweep", "Madhya Pradesh", "Maharashtra",
    "Manipur", "Meghalaya", "Mizoram", "Nagaland", "Odisha", "Puducherry", "Punjab",
    "Rajasthan", "Sikkim", "Tamil Nadu", "Telangana", "Tripura", "Uttar Pradesh",
    "Uttarakhand", "West Bengal"
]

# Create a DataFrame with the 'States' and 'Population' columns
df = pd.DataFrame({'States': indian_states})

# Define the initial population for 2020 (example)
initial_population_2020 = [417036, 53903393, 1570458, 35607039, 124799926, 1158473, 29436231,
    615724, 18710922, 1586250, 63872399, 28204692, 7451955, 13606320, 38593948,
    67562686, 35699443, 289023, 73183, 85358965, 123144223, 3091545, 3366710,
    1239244, 2249695, 46356334, 1413542, 30141373, 81032689, 690251, 77841267,
    38510982, 4169794, 237882725, 11250858, 99609303]

df['Population'] = initial_population_2020

# Define the calculate_vehicles function
def calculate_vehicles(population, year):
    # Start with the number of vehicles based on population
    vehicles = population // 10000  # One vehicle for every 1000 people (example)

    # Apply random variations to simulate increases (positive variations)
    # You can adjust the range (0% to +10%) to control the randomness
    variation = random.uniform(0.1, 0.4)

    # Calculate the new number of vehicles, ensuring it's always greater than the previous year
    if year == 2020:
        return vehicles
    else:
        previous_year_vehicles = df.at[df.index[df['Population'] == population].item(), f'No of Vehicles {year - 1}']
        new_vehicles = (int(vehicles * variation)+previous_year_vehicles)
        return new_vehicles

# Define the years for which you want to calculate the number of vehicles
years = [2020, 2021, 2022, 2023]

# Apply the calculate_vehicles function to calculate the number of vehicles for each year
for year in years:
    df[f'No of Vehicles {year}'] = df.apply(lambda row: calculate_vehicles(row['Population'], year), axis=1)

# Now, 'df' contains columns for each year's 'No of Vehicles' with incremental growth.
```
